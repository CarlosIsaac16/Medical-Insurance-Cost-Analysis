When dealing with missing values these are our four options:
1) If access to the business users who maintain the data, investigate what the missing value should be.


2) Sometimes it is possible to infer what the missing value should be.


3) If the variable is not a critical one, for instance is not the "y" / target variable, we can fill out:
For categorical variables, use the mode.
For numerical variables, use average or median.


4) Last resort is to delete the record altogether.

# A4.c Scatter Plots and Adding Features Medical Bills


Using scatter plots to identify outliers

At the end of the last section of this guide, we removed the outliers based on the descriptive statistics values. However, some values may be outliers in the context of other variables. For example, weight and height must be correlated.
A person who is tall is, on average, heavier than a person who is short.
As we last explored, there is a datapoint of a person who is >400 pounds heavy. Let's look at all subjects of the Medical Bills dataset with their weights plotted against their heights.

```python
import matplotlib.pyplot as plp

# TO MAKE THE CODE MORE LEGIBLE, WE WILL REFER TO THE
# `medical_bills_df` AS `df` ONLY FROM HERE ONWARDS
df = medical_bills_df

plp.plot(df['Height (cm)'], df['Weight (lbs)'], 'o', color='blue');
plp.xlabel("Height (cm)")
plp.ylabel('Weight (lbs)');
plp.show()
```

When plotting weight against height, we can observe a pattern: The taller the person the heavier, except for that datapoint which falls out of the pattern.

## Slicing the DataFrame according to specified criteria

Let’s slice the data to find the records whose weight is larger than 400 lb.

```python
df[df['Weight (lbs)'] > 400]
```

There are two individuals in the dataset whose weight is above 400 pounds. One is 187cm tall, and the other one is 140cm. While intuitively there seems to be something odd behind datapoint "864300", we need more data.

## Adding features  based of existing columns

Having domain knowledge is important whenever doing data analysis, in this case we know that BMI is a helpful measure that combines both height and weight. Calculating the BMI and running descriptive statistics on that new feature, will help us to determine whether that datapoint in question is an outlier or not.

```python
df["Weight (kg)"] = df["Weight (lbs)"] / 2.205
df["BMI"] = df["Weight (kg)"] / ( (df["Height (cm)"] / 100) ** 2 )
df.head()
```

We now have two columns in our dataset, we have the weight in KG's and we have the BMI measure.

## Running descriptive statistics on BMI

We re-run the summary statistics, and we look again at the records with weight above 400 lbs.

```python
df.describe()
df[df['Weight (lbs)'] > 400]
```

We observe the following about the datapoint in question:

3x higher than the mean.
9.7 standard deviations away from the mean.

It is safe to conclude that there is an data input issue on this datapoint, if we could, we would consult with the data owner. Sinc we can't, we are going to drop the datapoint.

# Regression in Python

Business Question

Could a company/government lower its insurance costs for employees/citizens should they drop weight?  

```python
df.head()
```

To avoid multicollinearity, I kept only one weight variable.

Since BMI already summarizes height and weight, I dropped both weight and height columns from my regression model

```python
# Remove columns related to weight and height (if still present)
df = df.drop(['Weight (lbs)', 'Weight (kg)', 'Height (cm)'], axis=1, errors='ignore')
```

```python
df.head()
```

## Use get_dummies

Regressions can only use numbers, so I converted all text columns, like gender or region, into numbers using the pandas get_dummies function.

```python
df = pd.get_dummies(df, drop_first=True)
```

## Non-linear Features

Sometimes, relationships are not just straight lines. To test for non-linear relationships,

I created new columns: BMI squared and Age squared.

```python
df['BMI_squared'] = df['BMI'] ** 2
df['Age_squared'] = df['Age'] ** 2
```

## Set X (features) and y (target)

The variable I want to predict is Annual Expenditure in Health Care. That’s my y variable.

Everything else is X.

```python
y = df['Annual Expenditure in HealthCare']
X = df.drop('Annual Expenditure in HealthCare', axis=1)
```

## Run the Regression Model

Now I can run my regression.

I use scikit-learn's LinearRegression model to fit the data and get the R² value and coefficients

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)

print("R^2:", model.score(X, y))
coef = pd.Series(model.coef_, index=X.columns)
print(coef)
```

The R² value is 1.0, which means my model fits the data perfectly.

The coefficient on Age means that spending decreases slightly with age. With each child born, costs increase.

A high BMI, or being a smoker, significantly increases medical expenses.

The quadratic effect of BMI and Age shows that extremes affect costs slightly. Also, being male or living in the Southwest slightly reduces expenses, according to my model.

## Plot Non-linear Relationships

To visualize the relationship,

I plot scatter plots for BMI and BMI squared against the target variable

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.scatter(df['BMI'], y, alpha=0.5)
plt.xlabel('BMI')
plt.ylabel('Annual Expenditure in Health Care')
plt.title('BMI vs Annual Expenditure')
plt.show()

plt.figure(figsize=(8,5))
plt.scatter(df['BMI_squared'], y, alpha=0.5)
plt.xlabel('BMI squared')
plt.ylabel('Annual Expenditure in Health Care')
plt.title('BMI² vs Annual Expenditure')
plt.show()
```

According to my regression, BMI has a positive and significant effect on annual health care expenditure.

This means that, in general, if people lose weight (lower BMI), their expected health care costs decrease.


Other factors that also affect costs are age, smoking, and possibly region or gender, depending on their coefficient and p-value.

```python
# X and y for BMI only
X_bmi = df[['BMI']] # Make sure it's a DataFrame, not a Series
y = df['Annual Expenditure in HealthCare']

# Fit the simple model with BMI only
model_bmi = LinearRegression()
model_bmi.fit(X_bmi, y)

# Predictions to plot the line
y_pred = model_bmi.predict(X_bmi)

# Plot with regression line
plt.figure(figsize=(8,5))
plt.scatter(df['BMI'], y, alpha=0.5, label='Data points')
plt.plot(df['BMI'], y_pred, color='red', label='Regression line')
plt.xlabel('BMI')
plt.ylabel('Annual Expenditure in HealthCare')
plt.title('BMI vs Annual Expenditure with Regression Line')
plt.legend()
plt.show()
```

We can see that as BMI increases, the predicted health expenditure also increases.

**Which other factors, cofounders, affect insurance cost?**

being a smoker greatly increases costs, having more children raises costs, and in my model, gender and living in certain regions can slightly reduce or increase the expected health care expenditure
